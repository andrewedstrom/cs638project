```python
#combine country columns
# author: @andrewedstrom

import operator, re, string

country_column = []
country_count = []

#create delimiter tokenizer using semicolon as delimiter
semi_tok = sm.DelimiterTokenizer(delim_set=['; '], return_set=True)
comma_tok = sm.DelimiterTokenizer(delim_set=[', '], return_set=True)

def clean_tokens(tok_list):
    new_list = []
    for s in tok_list:
        s = s.lower()
        
        #parentheticals
        if '(' in s:
            pstart = s.find('(')
            pend = s.find(')')
            if pend < len(s) - 1 and pend > 0:
                s = s[0:pstart] + s[pend+1:]
            else:
                s = s[0:pstart]
                
        #leading/trailing whitespace
        s = s.strip()
        
        #leading/trailing punctuation
        s = s.strip(string.punctuation)
        s = re.sub(r'[^a-zA-Z0-9 ]',r'',s)

        new_list.append(s)
    return new_list

for index, row in df.iterrows():
    lc = row['redlist_countries']
    rc = row['arkive_countries']
    
    ltok = semi_tok.tokenize(lc)
    rtok = comma_tok.tokenize(rc)
    ltok = clean_tokens(ltok)
    rtok = clean_tokens(rtok)
    
    d = {}
    
    for t in ltok:
        d[t] = True
    for t in rtok:
        if not d.get(t):
            d[t] = True
    keys = [k for k in d.keys()]
    
    
    country_list = ""
    for k in keys: 
        country_list += k + "; "
    country_list = country_list.strip().strip(string.punctuation)
    print(country_list)
    print()
    
    count = len(d)
    country_column.append(country_list)
    country_count.append(count)

df['countries'] = country_column
df['country_count'] = country_count
df.head()
```

```python
#drop  arkive and redlist specific country and country count columns

df.drop('redlist_country_count', axis=1, inplace=True)
df.drop('redlist_countries', axis=1, inplace=True)
df.drop('arkive_country_count', axis=1, inplace=True)
df.drop('arkive_countries', axis=1, inplace=True)
```

```python
#combine name columns
# author: @andrewedstrom

import operator, re, string

name_column = []
comma_tok = sm.DelimiterTokenizer(delim_set=[', '], return_set=True)

for index, row in df.iterrows():
    lc = row['redlist_name']
    rc = row['arkive_name']
    if not (type(lc) is str or type(rc) is str):
        name_column.append(None)
        continue
    
    ltok = None
    rtok = None
    if type(lc) is str:
        ltok = comma_tok.tokenize(lc)
        ltok = clean_tokens(ltok)

    if type(rc) is str:
        rtok = comma_tok.tokenize(rc)
        rtok = clean_tokens(rtok)
    
    d = {}
    
    if ltok:
        for t in ltok:
            d[t] = True
    if rtok:
        for t in rtok:
            if not d.get(t):
                d[t] = True
    keys = [k for k in d.keys()]
    
    
    name_list = ""
    for k in keys: 
        name_list += k + ", "
    name_list = name_list.strip().strip(string.punctuation)
    print(name_list)
    
    name_column.append(name_list)

df['names'] = name_column
df.head()
```
